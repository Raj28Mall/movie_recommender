```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import json
from collections import defaultdict
```

```python
movies_data=pd.read_csv("tmdb_5000_movies.csv")
movie_genres=movies_data["genres"]
from collections import defaultdict
genre_count = defaultdict(int)
for genre_list_str in movie_genres:
    genre_list = json.loads(genre_list_str)
    for genre in genre_list:
        genre_count[genre['name']] += 1
```

```python
genres= genre_count.keys()
genre_counts= np.array(list(genre_count.values()))
plt.figure(figsize=(10, 6))
plt.bar(genres, genre_counts, color='skyblue')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movie Genres')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
